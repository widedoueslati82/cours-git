# Partie 2 - Techniques de prétraitement de base

### 2.1 Traitement des données catégoriques

Question 1: Trouver quelles sont les colonnes de type catégoriques (textuelles) dans les données

```python
# Votre code ici
df.select_dtypes(include=["object", "category"]).columns
```

Question 2: Identifier les colonnes qui devraient être enlevées des données (Indice: Est-ce que toutes les colonnes catégoriques que vous avez identifiées sont pertinente afin de prédire si un passager a survécu ou non?)

```python
# Votre code ici
df = df.drop(columns=['Name', 'Ticket', 'Cabin'])
```

```python
df
```

Question 3: Identifier les colonnes qui devraient être encodées avec un One-Hot Encoding et faite l'encodage

```python
# Votre code ici
df = pd.get_dummies(df, columns=['Sex', 'Embarked'])
```

```python
df
```

### 2.2 Gestion des données manquantes

Question 4: Trouver le nombre de données manquantes par colonne dans les données

```python
# Votre code ici
df.isnull().sum()
```

Question 5: Éliminer les données manquantes avec la méthode la plus appropriée

```python
# Votre code ici
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df
```
